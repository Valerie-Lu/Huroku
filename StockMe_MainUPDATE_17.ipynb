> Welcome to Stock Me! Introducing a bot to keep you up-to-date with **different stocks, their prices, ups and downs, and values**. With only a few clicks, stock data of several stocks in Dow Jones will be ready for you to use!

The bot features Google Sheets as its starting point, calling data using *Google Finance*. *Google APIs* allow the transfer of information into *Python* for data cleaning and manipulation. Here, the data is converted into callable iterables with several functions in place to use them. 

Once the data is deliverable, the telebot module allows us to send specified details to users. They see the information they are looking for and won't worry about having outdated data on-the-go!

Stock Me advocates for **financial literacy**, specifically knowledge about the stock market. To better achieve this, Stock Me also includes ***Stock Simulator 1.0*** that allows users to **buy and sell stocks** without the real risks, actively engaging and educating them at the same time!

Apart from its advocacy, the project was also inspired by the coding backgrounds and experiences of five college students who wanted to make their own interpretation of a stock bot.
